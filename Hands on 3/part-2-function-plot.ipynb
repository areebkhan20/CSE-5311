```python
import time
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
def f(n):
    x = 1
    for i in range(1, n):
        for j in range(1, n):
            x = x + 1
    return x
```

```python
# time of the function from 1-10,000 with 100 steps in between
n_values = np.arange(1, 10000, 100)
times = []

for n in n_values:
    start_time = time.time()
    f(n)
    end_time = time.time()
    times.append(end_time - start_time)
```

```python
# Line of best fit
def quadratic_fit(n, a, b, c):
    return a * n**2 + b * n + c

params, _ = curve_fit(quadratic_fit, n_values, times)
```

```python
# Plot results
plt.figure(figsize=(8, 5))
plt.scatter(n_values, times, label="Measured Time", color="blue")
plt.plot(n_values, quadratic_fit(n_values, *params), label="Quadratic Fit", color="red")
plt.xlabel("n")
plt.ylabel("Execution Time (s)")
plt.title("Time Complexity of f(n)")
plt.legend()
plt.show()
```
